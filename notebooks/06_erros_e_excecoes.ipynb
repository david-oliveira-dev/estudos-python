# Notebook 06 — Erros e Exceções

Programas robustos não quebram quando algo dá errado: eles **tratam** o erro.

## Conteúdo
- tipos de erro;
- `try` / `except`;
- `else` e `finally`;
- capturar a mensagem do erro;
- `raise` (gerar erros);
- exceções personalizadas.

## 1. O que é uma exceção
Quando o Python encontra um problema em tempo de execução, ele lança uma
**exceção** e para. Exemplos comuns: `ValueError`, `ZeroDivisionError`,
`KeyError`, `FileNotFoundError`.

```python
# Sem tratamento, isto quebraria o programa:
try:
    resultado = 10 / 0
except ZeroDivisionError:
    print("Não dá para dividir por zero.")
```

## 2. try / except
O `try` tem o código que pode falhar; o `except` reage ao erro.

```python
def converter_para_int(texto):
    try:
        return int(texto)
    except ValueError:
        print(f"'{texto}' não é um número válido.")
        return None

print(converter_para_int("42"))
print(converter_para_int("abc"))
```

## 3. Capturando a mensagem do erro com `as`

```python
try:
    numeros = [1, 2, 3]
    print(numeros[10])
except IndexError as erro:
    print("Deu erro:", erro)
```

## 4. Vários except
Cada tipo de erro pode ter um tratamento diferente.

```python
def dividir(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "erro: divisão por zero"
    except TypeError:
        return "erro: tipos inválidos"

print(dividir(10, 2))
print(dividir(10, 0))
print(dividir(10, "x"))
```

## 5. else e finally
- `else`: roda se **não** deu erro;
- `finally`: roda **sempre** (deu erro ou não). Bom para "fechar" recursos.

```python
try:
    valor = int("100")
except ValueError:
    print("falhou")
else:
    print("deu certo, valor =", valor)
finally:
    print("fim do bloco (sempre executa)")
```

## 6. Gerando erros com raise
Eu mesmo posso sinalizar que um dado é inválido.

```python
def definir_idade(idade):
    if idade < 0:
        raise ValueError("idade não pode ser negativa")
    return idade

try:
    definir_idade(-5)
except ValueError as erro:
    print("recusado:", erro)
```

## 7. Exceções personalizadas
Posso criar minha própria exceção herdando de `Exception`. Deixa o código mais
expressivo.

```python
class SaldoInsuficienteError(Exception):
    pass

def sacar(saldo, valor):
    if valor > saldo:
        raise SaldoInsuficienteError(f"saldo R$ {saldo} é menor que R$ {valor}")
    return saldo - valor

try:
    sacar(100, 150)
except SaldoInsuficienteError as erro:
    print("operação negada:", erro)
```
