# Notebook 09 — Arquivos: Texto, CSV e JSON

Programas úteis precisam **guardar** e **ler** dados. Aqui vejo os três formatos
mais comuns. Todos usam apenas a biblioteca padrão.

## Conteúdo
- abrir arquivos com `with`;
- escrever e ler texto;
- CSV com o módulo `csv`;
- JSON com o módulo `json`.

## 1. Escrevendo um arquivo de texto
O `with` abre o arquivo e o fecha automaticamente no fim. Modo `"w"` = escrever
(apaga o conteúdo anterior).

```python
with open("exemplo.txt", "w", encoding="utf-8") as f:
    f.write("primeira linha\n")
    f.write("segunda linha\n")

print("arquivo escrito")
```

## 2. Lendo um arquivo de texto

```python
with open("exemplo.txt", "r", encoding="utf-8") as f:
    conteudo = f.read()
print(conteudo)

# linha por linha:
with open("exemplo.txt", "r", encoding="utf-8") as f:
    for linha in f:
        print("->", linha.strip())
```

## 3. Modo "a" (append)
Adiciona ao fim sem apagar o que já existe.

```python
with open("exemplo.txt", "a", encoding="utf-8") as f:
    f.write("linha adicionada depois\n")

with open("exemplo.txt", "r", encoding="utf-8") as f:
    print(f.read())
```

## 4. CSV (planilhas)
O módulo `csv` lê e escreve dados em formato de tabela (colunas separadas por
vírgula). `DictWriter`/`DictReader` trabalham com dicionários.

```python
import csv

alunos = [
    {"nome": "Ana", "media": 8.5},
    {"nome": "João", "media": 6.0},
]

with open("alunos.csv", "w", newline="", encoding="utf-8") as f:
    escritor = csv.DictWriter(f, fieldnames=["nome", "media"])
    escritor.writeheader()
    escritor.writerows(alunos)

print("CSV escrito")
```

```python
import csv

with open("alunos.csv", "r", encoding="utf-8") as f:
    leitor = csv.DictReader(f)
    for linha in leitor:
        print(linha["nome"], "->", linha["media"])
```

## 5. JSON (o formato das APIs)
JSON é o formato mais usado para trocar dados na web. O módulo `json` converte
entre objetos Python (dicionários/listas) e texto JSON.

```python
import json

dados = {
    "nome": "David",
    "linguagens": ["Python", "SQL"],
    "ativo": True,
}

# objeto Python -> texto JSON (salvar)
with open("perfil.json", "w", encoding="utf-8") as f:
    json.dump(dados, f, ensure_ascii=False, indent=4)

print("JSON salvo")
```

```python
import json

# texto JSON -> objeto Python (carregar)
with open("perfil.json", "r", encoding="utf-8") as f:
    carregado = json.load(f)

print(carregado["nome"])
print(carregado["linguagens"])
print(type(carregado))
```

## 6. JSON a partir de uma string
Útil quando recebo a resposta de uma API (que vem como texto).

```python
import json

resposta = '{"cidade": "Recife", "temperatura": 29.5}'
dados = json.loads(resposta)     # loads = "load string"
print(dados["cidade"], "-", dados["temperatura"], "°C")
```
