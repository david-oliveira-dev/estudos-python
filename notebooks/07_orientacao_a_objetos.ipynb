# Notebook 07 — Orientação a Objetos (POO)

POO junta **dados** (atributos) e **comportamentos** (métodos) dentro de classes.

## Conteúdo
- classe e objeto;
- `__init__` e `self`;
- métodos;
- `__str__` e outros dunder;
- encapsulamento;
- herança.

## 1. Primeira classe
A classe é o molde; o objeto é cada instância criada a partir dela.

```python
class Cachorro:
    def __init__(self, nome, raca):
        self.nome = nome      # atributo do objeto
        self.raca = raca

    def latir(self):          # método
        return f"{self.nome} faz: Au au!"

rex = Cachorro("Rex", "Labrador")
print(rex.nome, "-", rex.raca)
print(rex.latir())
```

## 2. O método __str__
Controla como o objeto aparece em `print`.

```python
class Produto:
    def __init__(self, nome, preco):
        self.nome = nome
        self.preco = preco

    def __str__(self):
        return f"{self.nome}: R$ {self.preco:.2f}"

print(Produto("Caderno", 12.9))
```

## 3. Métodos que usam os atributos

```python
class ContaBancaria:
    def __init__(self, titular, saldo=0.0):
        self.titular = titular
        self.saldo = saldo

    def depositar(self, valor):
        self.saldo += valor

    def sacar(self, valor):
        if valor > self.saldo:
            return "saldo insuficiente"
        self.saldo -= valor
        return "saque realizado"

    def __str__(self):
        return f"Conta de {self.titular} | saldo: R$ {self.saldo:.2f}"

conta = ContaBancaria("David", 100)
conta.depositar(50)
print(conta.sacar(30))
print(conta)
```

## 4. Encapsulamento
Por convenção, um atributo começando com `_` é "interno" (não deve ser mexido de
fora diretamente). Usamos métodos para controlar o acesso.

```python
class Termometro:
    def __init__(self):
        self._celsius = 0

    def definir(self, valor):
        if valor < -273.15:
            raise ValueError("abaixo do zero absoluto")
        self._celsius = valor

    def em_fahrenheit(self):
        return self._celsius * 9 / 5 + 32

t = Termometro()
t.definir(25)
print(t.em_fahrenheit())
```

## 5. Herança
Uma classe pode herdar de outra, reaproveitando seus atributos e métodos, e
acrescentando ou trocando o que precisar. `super()` chama a classe-mãe.

```python
class Animal:
    def __init__(self, nome):
        self.nome = nome

    def descrever(self):
        return f"{self.nome} é um animal."

class Gato(Animal):
    def __init__(self, nome, cor):
        super().__init__(nome)    # reusa o __init__ da classe Animal
        self.cor = cor

    def descrever(self):          # sobrescreve o método
        return f"{self.nome} é um gato {self.cor}."

print(Animal("Bicho").descrever())
print(Gato("Mimi", "preto").descrever())
```

## 6. Exemplo prático: lista de objetos
POO brilha quando temos vários objetos do mesmo tipo.

```python
class Aluno:
    def __init__(self, nome, notas):
        self.nome = nome
        self.notas = notas

    def media(self):
        return sum(self.notas) / len(self.notas)

turma = [
    Aluno("Ana", [8.0, 9.0, 7.0]),
    Aluno("João", [5.0, 6.0, 4.0]),
]

for aluno in turma:
    status = "aprovado" if aluno.media() >= 7 else "recuperação"
    print(f"{aluno.nome}: média {aluno.media():.1f} ({status})")
```
