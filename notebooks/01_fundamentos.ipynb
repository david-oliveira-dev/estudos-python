## 1. print e comentários
O `print` mostra algo na tela. O `#` cria um comentário (o Python ignora).

```python
# isto é um comentário — não é executado
print("Olá, mundo!")
print("Estou estudando Python.")
```

## 2. Variáveis
Uma variável é um nome que aponta para um valor guardado na memória.

```python
nome = "David"
idade = 24
altura = 1.75

print(nome, idade, altura)
```

## 3. Tipos de dados
Os tipos básicos do Python:
- `int` — números inteiros (24);
- `float` — números com casas decimais (1.75);
- `str` — texto (entre aspas);
- `bool` — verdadeiro/falso (`True` / `False`).

A função `type()` mostra o tipo de um valor.

```python
print(type(24))      # int
print(type(1.75))    # float
print(type("David")) # str
print(type(True))    # bool
```

## 4. Operadores aritméticos

```python
a = 10
b = 3

print("soma:", a + b)
print("subtração:", a - b)
print("multiplicação:", a * b)
print("divisão:", a / b)        # sempre devolve float
print("divisão inteira:", a // b)
print("resto (módulo):", a % b)
print("potência:", a ** b)
```

## 5. Operadores de comparação
Devolvem sempre um `bool` (`True` ou `False`).

```python
print(10 > 3)
print(10 == 3)
print(10 != 3)
print(10 <= 10)
```

## 6. Operadores lógicos
`and`, `or` e `not` combinam condições.

```python
tem_idade = idade >= 18
tem_documento = True

print(tem_idade and tem_documento)
print(tem_idade or False)
print(not tem_idade)
```

## 7. Conversão de tipos (casting)
Às vezes preciso transformar um tipo em outro: texto em número, número em texto.

```python
texto = "100"
numero = int(texto)      # str -> int
print(numero + 50)

preco = 19.9
print("R$ " + str(preco))  # float -> str para juntar com texto
```

## 8. Entrada de dados com input
O `input()` lê o que o usuário digita. **Atenção:** ele sempre devolve `str`,
então para fazer contas eu preciso converter.

> Em um notebook, o `input` abre uma caixinha. Por isso, abaixo simulo a entrada
> com uma variável, para o notebook rodar sozinho.

```python
# Em um programa de verdade seria:
# nome = input("Qual seu nome? ")
# idade = int(input("Qual sua idade? "))

# Aqui simulo a entrada já convertida:
nome_digitado = "David"
idade_digitada = int("24")

print(f"Olá, {nome_digitado}! No ano que vem você terá {idade_digitada + 1} anos.")
```
