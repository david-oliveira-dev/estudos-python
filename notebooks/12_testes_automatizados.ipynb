Testes verificam, de forma automática, se o código faz o que deveria. Eles dão
segurança para mudar o código sem medo de quebrar algo.

## Conteúdo
- por que testar;
- `assert`;

- testando se um erro acontece.

## 1. A ideia mais simples: assert
`assert` verifica se algo é verdadeiro. Se não for, dá erro.

```python
def somar(a, b):
    return a + b

assert somar(2, 3) == 5
assert somar(-1, 1) == 0
print("todos os asserts passaram")
```

## 2. unittest — a forma profissional
Vem na biblioteca padrão. Organizamos os testes em uma classe que herda de
`unittest.TestCase`, e cada método começando com `test_` é um teste.

```python
import unittest

# função que queremos testar
def eh_par(n):
    return n % 2 == 0


class TestEhPar(unittest.TestCase):
    def test_numero_par(self):
        self.assertTrue(eh_par(4))

    def test_numero_impar(self):
        self.assertFalse(eh_par(7))

    def test_zero(self):
        self.assertTrue(eh_par(0))


# Em um arquivo normal usaríamos: unittest.main()
# No notebook, rodamos assim para não conflitar com o Jupyter:
unittest.main(argv=[""], exit=False, verbosity=2)
print("fim dos testes")
```

## 3. Os principais asserts
- `assertEqual(a, b)` — a == b
- `assertTrue(x)` / `assertFalse(x)`
- `assertIn(item, colecao)`
- `assertRaises(Erro)` — verifica se um erro é lançado

```python
import unittest

def dividir(a, b):
    if b == 0:
        raise ValueError("divisão por zero")
    return a / b


class TestDividir(unittest.TestCase):
    def test_divisao_normal(self):
        self.assertEqual(dividir(10, 2), 5)

    def test_resultado_float(self):
        self.assertAlmostEqual(dividir(10, 3), 3.3333, places=3)

    def test_divisao_por_zero_gera_erro(self):
        with self.assertRaises(ValueError):
            dividir(10, 0)


unittest.main(argv=[""], exit=False, verbosity=2)
print("fim dos testes")
```

## 4. setUp — preparando cada teste
O método `setUp` roda **antes de cada** teste, para evitar repetição.

```python
import unittest

class Carrinho:
    def __init__(self):
        self.itens = []
    def adicionar(self, item):
        self.itens.append(item)
    def total(self):
        return len(self.itens)


class TestCarrinho(unittest.TestCase):
    def setUp(self):
        self.carrinho = Carrinho()        # carrinho novo a cada teste

    def test_comeca_vazio(self):
        self.assertEqual(self.carrinho.total(), 0)

    def test_adicionar_item(self):
        self.carrinho.adicionar("livro")
        self.assertEqual(self.carrinho.total(), 1)


unittest.main(argv=[""], exit=False, verbosity=2)
print("fim dos testes")
```
