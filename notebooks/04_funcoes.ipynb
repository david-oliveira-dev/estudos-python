# Notebook 04 — Funções

Funções organizam o código em blocos reutilizáveis, evitando repetição.

## Conteúdo
- `def` e `return`;
- parâmetros e argumentos;
- valores padrão e argumentos nomeados;
- `*args` e `**kwargs`;
- escopo (local vs global);
- funções `lambda`;
- type hints (dicas de tipo).

## 1. Definindo uma função
`def` cria a função; `return` devolve um resultado.

```python
def saudacao(nome):
    return f"Olá, {nome}!"

print(saudacao("David"))
print(saudacao("Ana"))
```

## 2. Vários parâmetros e valor padrão

```python
def calcular_total(preco, quantidade, desconto=0.0):
    total = preco * quantidade
    return total - (total * desconto)

print(calcular_total(10, 3))            # sem desconto
print(calcular_total(10, 3, 0.1))       # 10% de desconto
print(calcular_total(preco=10, quantidade=3, desconto=0.5))  # argumentos nomeados
```

## 3. *args e **kwargs
- `*args` recebe vários argumentos posicionais como uma **tupla**;
- `**kwargs` recebe vários argumentos nomeados como um **dicionário**.

```python
def somar_tudo(*numeros):
    return sum(numeros)

print(somar_tudo(1, 2, 3, 4, 5))


def descrever(**dados):
    for chave, valor in dados.items():
        print(f"{chave}: {valor}")

descrever(nome="Ana", idade=15, cidade="Recife")
```

## 4. Escopo de variáveis
Variáveis criadas dentro de uma função são **locais** — só existem ali dentro.

```python
mensagem = "global"

def teste():
    mensagem = "local"   # esta é outra variável, só existe aqui
    print("dentro:", mensagem)

teste()
print("fora:", mensagem)
```

## 5. Funções lambda
Funções pequenas e anônimas, em uma linha. Muito usadas com `sorted`, `map`, etc.

```python
dobro = lambda x: x * 2
print(dobro(5))

alunos = [{"nome": "Ana", "media": 8.5}, {"nome": "João", "media": 6.0}]
# ordena pela média, da maior para a menor
ordenados = sorted(alunos, key=lambda a: a["media"], reverse=True)
print(ordenados)
```

## 6. Type hints (dicas de tipo)
Anotações que indicam os tipos esperados. Não mudam a execução, mas deixam o
código mais claro e ajudam editores e ferramentas.

```python
def media(notas: list[float]) -> float:
    return sum(notas) / len(notas)

print(media([8.0, 7.5, 9.0]))
```

## 7. Documentando com docstring
A primeira string dentro da função descreve o que ela faz.

```python
def fatorial(n: int) -> int:
    """Calcula o fatorial de n (n!). Ex.: fatorial(5) = 120."""
    resultado = 1
    for i in range(2, n + 1):
        resultado *= i
    return resultado

print(fatorial(5))
print(fatorial.__doc__)
```
