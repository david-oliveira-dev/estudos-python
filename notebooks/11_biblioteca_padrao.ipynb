## 1. datetime — datas e horas

```python
from datetime import datetime, date, timedelta

agora = datetime.now()
print("agora:", agora.strftime("%d/%m/%Y %H:%M"))

nascimento = date(2002, 5, 20)
hoje = date.today()
idade_em_dias = (hoje - nascimento).days
print("dias vividos:", idade_em_dias)

daqui_uma_semana = hoje + timedelta(days=7)
print("daqui uma semana:", daqui_uma_semana.strftime("%d/%m/%Y"))
```

## 2. math — matemática

```python
import math

print(math.sqrt(2))
print(math.floor(4.7), math.ceil(4.2))
print(math.gcd(12, 18))        # máximo divisor comum
print(round(math.pi, 4))
```

## 3. random — aleatoriedade

```python
import random

random.seed(42)   # deixa o resultado reproduzível (bom para testar)

print(random.randint(1, 100))
print(random.choice(["pedra", "papel", "tesoura"]))

baralho = list(range(1, 11))
random.shuffle(baralho)
print(baralho)

print(random.sample(range(1, 61), 6))   # 6 dezenas únicas (tipo mega-sena)
```

## 4. collections — estruturas extras
- `Counter`: conta ocorrências;
- `defaultdict`: dicionário com valor padrão.

```python
from collections import Counter, defaultdict

texto = "banana"
print(Counter(texto))     # conta cada letra

votos = ["A", "B", "A", "C", "A", "B"]
print(Counter(votos).most_common(1))   # mais votado

# agrupando com defaultdict
alunos = [("9A", "Ana"), ("9B", "João"), ("9A", "Maria")]
turmas = defaultdict(list)
for turma, nome in alunos:
    turmas[turma].append(nome)
print(dict(turmas))
```

## 5. pathlib — caminhos de arquivos
Forma moderna de lidar com arquivos e pastas (melhor que juntar strings).

```python
from pathlib import Path

caminho = Path("dados") / "relatorio.txt"
print("caminho:", caminho)
print("nome do arquivo:", caminho.name)
print("extensão:", caminho.suffix)
print("pasta-mãe:", caminho.parent)

# o diretório atual e os .py dentro de scripts/
atual = Path.cwd()
print("pasta atual existe?", atual.exists())
```
