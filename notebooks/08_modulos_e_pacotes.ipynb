## 1. Importando da biblioteca padrão
O Python já vem com centenas de módulos prontos.

````python
import math
import random

print(math.sqrt(144))
print(math.pi)
print(random.randint(1, 6))   # simula um dado
````

## 2. Formas de importar

````python
# importa o módulo inteiro
import math
print(math.factorial(5))

# importa só o que preciso
from math import sqrt, pi
print(sqrt(81), pi)

# importa com um apelido
import math as m
print(m.ceil(4.2))
````

## 3. Criando seu próprio módulo
Aqui crio um arquivo `calculadora.py` em tempo de execução, só para demonstrar.
Na prática, você criaria esse arquivo no editor.

````python
codigo = '''
def somar(a, b):
    return a + b

def subtrair(a, b):
    return a - b

PI = 3.14159
'''

with open("calculadora.py", "w", encoding="utf-8") as f:
    f.write(codigo)

print("módulo calculadora.py criado")
````

````python
import importlib
import calculadora
importlib.reload(calculadora)   # garante a versão mais recente no notebook

print(calculadora.somar(2, 3))
print(calculadora.subtrair(10, 4))
print(calculadora.PI)
````

## 4. O bloco `if __name__ == "__main__"`
Código dentro desse bloco só roda quando o arquivo é executado **diretamente**
(`python arquivo.py`), e **não** quando ele é importado. Serve para separar
"o que é biblioteca" de "o que é programa".

````python
exemplo = '''
def saudacao():
    return "olá do módulo"

if __name__ == "__main__":
    # só roda com: python exemplo.py
    print("executando direto:", saudacao())
'''
print(exemplo)
````

## 5. Pacotes
Um **pacote** é uma pasta com vários módulos. Estrutura típica de projeto:

```
meu_projeto/
├── meu_pacote/
│   ├── __init__.py     <- marca a pasta como pacote
│   ├── modelos.py
│   └── utils.py
└── main.py
```

Importação: `from meu_pacote.utils import alguma_funcao`.

## 6. pip e ambientes virtuais
Para instalar bibliotecas de terceiros usamos o **pip**. Para não misturar as
dependências de projetos diferentes, criamos um **ambiente virtual** (venv):

```bash
python3 -m venv .venv          # cria o ambiente
source .venv/bin/activate      # ativa (Linux/Mac)
pip install requests           # instala dentro do ambiente
pip freeze > requirements.txt  # salva a lista de dependências
```

Cada projeto deste repositório segue exatamente esse fluxo.
