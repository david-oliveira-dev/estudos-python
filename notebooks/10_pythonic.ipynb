## Conteúdo
- list/dict/set comprehensions;
- `enumerate` e `zip`;
- iteradores;
- geradores (`yield`);
- funções `map`, `filter`.

## 1. List comprehension
Cria uma lista em uma linha. Substitui um `for` + `append`.

```python
# jeito tradicional
quadrados = []
for n in range(1, 6):
    quadrados.append(n ** 2)
print(quadrados)

# jeito pythonic
quadrados = [n ** 2 for n in range(1, 6)]
print(quadrados)
```

## 2. Comprehension com condição (filtro)

```python
numeros = range(1, 11)
pares = [n for n in numeros if n % 2 == 0]
print(pares)

# com transformação + filtro
dobro_dos_impares = [n * 2 for n in numeros if n % 2 != 0]
print(dobro_dos_impares)
```

## 3. Dict e set comprehension

```python
# dicionário: numero -> quadrado
quadrados = {n: n ** 2 for n in range(1, 6)}
print(quadrados)

# set: primeiras letras (sem repetir)
palavras = ["python", "php", "java", "javascript"]
iniciais = {p[0] for p in palavras}
print(iniciais)
```

## 4. enumerate
Percorre uma sequência com o **índice** junto.

```python
linguagens = ["Python", "SQL", "JavaScript"]

for i, linguagem in enumerate(linguagens, start=1):
    print(f"{i}. {linguagem}")
```

## 5. zip
Percorre **duas (ou mais) sequências** em paralelo.

```python
nomes = ["Ana", "João", "Maria"]
notas = [8.5, 6.0, 9.2]

for nome, nota in zip(nomes, notas):
    print(f"{nome}: {nota}")

# zip também cria um dicionário facilmente
print(dict(zip(nomes, notas)))
```

## 6. Geradores (yield)
Um gerador produz valores **um de cada vez**, sob demanda, sem guardar tudo na
memória. Ótimo para sequências grandes.

```python
def contar_ate(n):
    numero = 1
    while numero <= n:
        yield numero      # "pausa" e entrega um valor por vez
        numero += 1

for valor in contar_ate(5):
    print(valor)

# expressão geradora (parecida com comprehension, mas com parênteses)
soma = sum(x ** 2 for x in range(1, 1001))
print("soma dos quadrados de 1 a 1000:", soma)
```

## 7. map e filter
Alternativas funcionais às comprehensions (hoje as comprehensions são preferidas,
mas é bom reconhecer).

```python
numeros = [1, 2, 3, 4, 5]

dobrados = list(map(lambda x: x * 2, numeros))
print(dobrados)

pares = list(filter(lambda x: x % 2 == 0, numeros))
print(pares)
```
