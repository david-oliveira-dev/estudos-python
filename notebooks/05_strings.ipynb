# Notebook 05 — Strings (Texto)

Texto é um dos tipos com que mais trabalhamos. Aqui vejo os métodos e truques
mais úteis.

## Conteúdo
- criando e juntando strings;
- métodos comuns (`upper`, `lower`, `strip`, `replace`, `split`...);
- fatiamento;
- f-strings e formatação de números;
- verificações úteis.

## 1. Criando e juntando

```python
nome = "David"
sobrenome = "Oliveira"

completo = nome + " " + sobrenome     # concatenação
print(completo)
print("ha" * 3)                       # repetição -> hahaha
print(len(completo))                  # tamanho
```

## 2. Métodos mais usados

```python
texto = "  Python é Demais  "

print(texto.upper())       # tudo maiúsculo
print(texto.lower())       # tudo minúsculo
print(texto.strip())       # remove espaços das pontas
print(texto.replace("Demais", "incrível"))
print(texto.strip().split(" "))   # quebra em lista
```

## 3. Fatiamento de strings
Funciona igual ao de listas: `texto[inicio:fim]`.

```python
palavra = "Python"
print(palavra[0])      # P
print(palavra[-1])     # n
print(palavra[0:3])    # Pyt
print(palavra[::-1])   # invertida -> nohtyP
```

## 4. f-strings (a melhor forma de formatar)
Coloca variáveis e expressões dentro do texto usando `{}`.

```python
produto = "café"
preco = 5.5
quantidade = 3

print(f"{quantidade}x {produto} = R$ {preco * quantidade:.2f}")
print(f"O dobro de 21 é {21 * 2}")
```

## 5. Formatando números

```python
valor = 1234.5678

print(f"2 casas decimais: {valor:.2f}")
print(f"separador de milhar: {valor:,.2f}")
print(f"porcentagem: {0.1567:.1%}")
print(f"preenchendo com zero: {7:03d}")
```

## 6. Verificações úteis
Úteis para validar entradas.

```python
print("123".isdigit())          # só números?
print("abc".isalpha())          # só letras?
print("python".startswith("py"))
print("arquivo.csv".endswith(".csv"))
print("Ana" in "Banana")        # contém?
```

## 7. Exemplo prático: limpar e padronizar nomes

```python
nomes_baguncados = ["  ana ", "JOÃO", "Maria  "]

limpos = [nome.strip().capitalize() for nome in nomes_baguncados]
print(limpos)
```
