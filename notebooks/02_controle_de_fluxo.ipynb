# Notebook 02 — Controle de Fluxo

Até agora os programas rodavam de cima para baixo, sempre igual. O **controle de
fluxo** permite tomar decisões e repetir ações.

## Conteúdo
- `if`, `elif`, `else`;
- operadores em condições;
- laço `while`;
- laço `for` e `range`;
- `break` e `continue`.

## 1. if / elif / else
Executa um bloco **só** quando a condição é verdadeira.

```python
nota = 7.5

if nota >= 9:
    print("Excelente!")
elif nota >= 7:
    print("Aprovado.")
elif nota >= 5:
    print("Recuperação.")
else:
    print("Reprovado.")
```

## 2. Condições combinadas
Dá para juntar condições com `and`, `or`, `not`.

```python
idade = 20
tem_carteira = True

if idade >= 18 and tem_carteira:
    print("Pode dirigir.")
else:
    print("Não pode dirigir.")
```

## 3. Laço while
Repete enquanto a condição for verdadeira. Cuidado para a condição **mudar**,
senão o laço nunca acaba (loop infinito).

```python
contador = 1
while contador <= 5:
    print("contando:", contador)
    contador += 1   # mesmo que contador = contador + 1

print("acabou")
```

## 4. Laço for e range
O `for` percorre uma sequência. O `range(n)` gera números de 0 até n-1.

```python
for i in range(5):
    print("i =", i)

print("---")

# range(inicio, fim, passo)
for par in range(0, 10, 2):
    print("par:", par)
```

## 5. Percorrendo uma lista com for

```python
linguagens = ["Python", "JavaScript", "SQL"]

for linguagem in linguagens:
    print("Estudando:", linguagem)
```

## 6. break e continue
- `break` interrompe o laço na hora;
- `continue` pula para a próxima volta.

```python
for numero in range(1, 11):
    if numero == 7:
        print("achei o 7, parando")
        break
    if numero % 2 == 0:
        continue   # pula os pares
    print("ímpar:", numero)
```

## 7. Exemplo prático: somar números
Juntando `for` + acumulador (um padrão que aparece muito).

```python
numeros = [10, 25, 30, 5]
total = 0

for n in numeros:
    total += n

print("soma total:", total)
print("média:", total / len(numeros))
```
